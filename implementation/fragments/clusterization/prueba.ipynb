```python
# Importing the necessary libraries for the data analysis and transformations
import pandas as pd
from joblib import load
```

```python
# Obtaining the dataframe to predict
train = pd.read_csv('../../../data/original_train.csv')

x_test = pd.read_csv('../../../data/test/x_test.csv')
y_test = pd.read_csv('../../../data/test/y_test.csv')
```

```python
x_train = train.drop(['TARGET'], axis=1)
y_train = pd.DataFrame(train['TARGET'])
```

```python
x_test = x_test.append(x_train)
y_test = y_test.append(y_train)
```

```python
x_test['TARGET'] = y_test
x_test = x_test.loc[x_test['TARGET'] == 1.0]
x_test = x_test.drop(['TARGET'], axis = 1)
y_test = y_test.loc[y_test['TARGET'] == 1.0]
y_test
```

```python
x_test
```

```python
# Loading the model|
model = load('../joblibs/classification-model-smote.joblib')
model
```

```python
#Predicting y for X_val
y_pred_prob = model.predict_proba(x_test)
y_pred = model.predict(x_test)
y_pred
```

```python
#Getting the accuracy of the model
model.score(x_test, y_test)
```
